```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('customer_satisfaction.csv')
df.head()
```

```python
branches = df['Branch']
satisfaction = df['Satisfaction Score']
```

```python
plt.figure(figsize=(10, 6))
plt.bar(branches, satisfaction, color='skyblue')
plt.title('Customer Satisfaction per Branch')
plt.xlabel('Branch')
plt.ylabel('Customer Satisfaction')
plt.ylim(88, 95)
plt.yticks(range(88, 95, 1))
plt.show()
```
